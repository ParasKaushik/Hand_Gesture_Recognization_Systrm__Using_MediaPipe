```python
# importing Libraries
import cv2
import numpy as np
import tensorflow as tf
import mediapipe as mp
from tensorflow.keras.models import load_model
```

```python
# Initlializing the meiapipe library
mpHands=mp.solutions.hands
hands=mpHands.Hands(max_num_hands=1,min_detection_confidence=0.7)
moDraw=mp.solutions.drawing_utils
```

```python
#oading the model
model=load_model('mp_hand_gesture')
```

```python
#LOad TArget Class NAmes
f=open('gesture.names','r')
classNames=f.read().split('\n')
f.close()
```

```python
classNames
```

```python
cap=cv2.VideoCapture(0)
while True:
    val,frame=cap.read()
    x,y,c=frame.shape
    #fliping the window
    frame=cv2.flip(frame,1)
    framegb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    result=hands.process(framegb)
    
    className=''
    if result.multi_hand_landmarks:
        landmarks=[]
        for h1 in result.multi_hand_landmarks:
            for lm in h1.landmark:
                lmx=(lm.x*x)
                lmy=(lm.y*y)
                
                landmarks.append([lmx,lmy])
            moDraw.draw_landmarks(frame,h1,mpHands.HAND_CONNECTIONS)
            #predict
            pred=model.predict([landmarks])
            classId=np.argmax(pred)
            className=classNames[classId]
            #put Text
    cv2.putText(frame,className,(1,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2,cv2.LINE_AA)
    #Showing Result
    cv2.imshow('Win',frame)
    if cv2.waitKey(1)&0xFF==ord('b'):
        break
cap.release()
cv2.destroyAllWindows()
```
